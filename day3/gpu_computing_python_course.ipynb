Let's import the packages

```python
import numpy as np
import cupy as cp
```

Let's perform the excercise

```python
x = np.random.random((128, 128))
y = np.random.random((256, 256))
z = np.random.random((512, 512))
v = np.random.random((1024, 1024))
w = np.random.random((2048, 2048))

%timeit np.fft.fft(x)
%timeit np.fft.fft(y)
%timeit np.fft.fft(z)
%timeit np.fft.fft(v)
%timeit np.fft.fft(w)
```

```python
x = cp.random.random((128, 128))
y = cp.random.random((256, 256))
z = cp.random.random((512, 512))
v = cp.random.random((1024, 1024))
w = cp.random.random((2048, 2048))

%timeit cp.fft.fft(x)
%timeit cp.fft.fft(y)
%timeit cp.fft.fft(z)
%timeit cp.fft.fft(v)
%timeit cp.fft.fft(w)
```

### At what sizes does cupy outperform?

The results shows that is faster since the 128x128 matrix, and it's about 2X. However, it start to have a biggest difference since the 512x512 matrix, which make it about 15x faster.

```python
x = np.random.random((128, 128))
x = x.astype(np.float32)
y = np.random.random((256, 256))
y = y.astype(np.float32)
z = np.random.random((512, 512))
z = z.astype(np.float32)
v = np.random.random((1024, 1024))
v = v.astype(np.float32)
w = np.random.random((2048, 2048))
w = w.astype(np.float32)

%timeit np.fft.fft(x)
%timeit np.fft.fft(y)
%timeit np.fft.fft(z)
%timeit np.fft.fft(v)
%timeit np.fft.fft(w)
```

```python
x = cp.random.random((128, 128))
x = x.astype(cp.float32)
y = cp.random.random((256, 256))
y = y.astype(cp.float32)
z = cp.random.random((512, 512))
z = z.astype(cp.float32)
v = cp.random.random((1024, 1024))
v = v.astype(cp.float32)
w = cp.random.random((2048, 2048))
w = w.astype(cp.float32)

%timeit cp.fft.fft(x)
%timeit cp.fft.fft(y)
%timeit cp.fft.fft(z)
%timeit cp.fft.fft(v)
%timeit cp.fft.fft(w)
```

### What happens now if you set the datatype of the array to numpy.float32?

In this case, the difference in time is bigger when we increased the matrix size. With float64 the difference in the biggest matrix was around 35x, while with float32 the difference was around 140x.
